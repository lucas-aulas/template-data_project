 **O que este script faz?**

 1. **Automação de Navegador (RPA/Scraping):** Abre um navegador real (Google Chrome) de forma automatizada.

 2. **Navegação Dinâmica:** Interage com a barra de pesquisa do site IMDB buscando por um ator específico ("Wagner Moura").

 3. **Mapeamento de Elementos HTML:** Localiza links através de classes CSS e clica no primeiro resultado para entrar no perfil do ator.

 4. **Engenharia de URLs (Data Parsing):** Captura a URL atual da página, quebra o texto (Split) e isola o ID único do ator no IMDB.

 5. **Navegação Direta e Captura:** Vai direto para a página de links externos do ator, localiza o link do Instagram usando caminhos XPATH e isola o nome de usuário da rede social.

```python
# ==============================================================================
# 1. IMPORTAÇÃO DE BIBLIOTECAS E DEPENDÊNCIAS
# ==============================================================================

# O Selenium é uma ferramenta robusta para automação de testes de software e web scraping.
# Ele simula um usuário real interagindo com um navegador (clicando, digitando, rolando a página).
from selenium import webdriver

# A classe 'By' é usada para definir a estratégia de localização de um elemento na página HTML.
# Permite encontrar botões, textos ou inputs por ID, nome da classe, tag HTML ou caminhos XPATH.
from selenium.webdriver.common.by import By

# A função 'sleep' suspende a execução do código por um número determinado de segundos.
# Essencial no scraping básico para esperar que o navegador carregue o conteúdo JavaScript da página.
from time import sleep
```

```python
# ==============================================================================
# 2. INICIALIZAÇÃO DO NAVEGADOR
# ==============================================================================

# Cria uma instância do navegador Google Chrome controlada pelo script.
# Uma janela do Chrome em branco se abrirá na sua tela de forma automatizada.
driver = webdriver.Chrome()

# Aguarda 5 segundos. Boa prática inicial para garantir que o software do navegador
# (Chromedriver) se conectou perfeitamente ao script Python antes de enviar comandos.
sleep(5)
```

```python
# ==============================================================================
# 3. NAVEGAÇÃO E PESQUISA AUTOMATIZADA
# ==============================================================================

# O método '.get()' instrui o navegador a acessar diretamente uma URL específica.
# Aqui, passamos a URL de busca do IMDB injetando dinamicamente o termo de pesquisa "Wagner Moura".
driver.get(f"https://www.imdb.com/pt/find/?q=Wagner Moura")

# Aguarda 5 segundos para que a página de resultados de busca carregue completamente
# todos os elementos de texto, imagens e links do ator.
sleep(5)
```

```python
# ==============================================================================
# 4. INTERAÇÃO COM ELEMENTOS DA PÁGINA (CLIQUE NO RESULTADO)
# ==============================================================================

# 'find_elements' (no plural) busca TODOS os elementos da página que usam a classe CSS especificada.
# Retorna uma lista do Python contendo esses elementos estruturados.
# '[0]' seleciona o primeiro elemento encontrado na lista (o resultado mais relevante da busca).
# '.click()' simula o clique do mouse do usuário em cima desse link/título do ator.
driver.find_elements(By.CLASS_NAME, "ipc-title-link-wrapper")[0].click()

# Aguarda 5 segundos para que a página de perfil biográfico do ator seja carregada.
sleep(5)
```

```python
# ==============================================================================
# 5. ENGENHARIA DE STRINGS E MAPEAMENTO DE ID (URL PARSING)
# ==============================================================================

# 'driver.current_url' captura o endereço web exato em que o navegador se encontra no momento.
# Exemplo de URL de perfil: "https://www.imdb.com/name/nm1074819/?ref_=fn_al_nm_1"
# '.split("/")' quebra essa string em partes toda vez que encontra uma barra ("/"), virando uma lista:
# ['https:', '', 'www.imdb.com', 'name', 'nm1074819', '?ref_=fn_al_nm_1']
# O índice '[5]' extrai especificamente a quinta posição da lista, que armazena o ID único do ator ('nm1074819').
act_id = driver.current_url.split("/")[5]

# Utiliza o ID extraído para construir uma nova URL padrão e navegar direto para a página
# de sites externos e redes sociais cadastradas daquele ator no IMDB.
driver.get(f"https://www.imdb.com/name/{act_id}/externalsites")

# Aguarda 5 segundos para o carregamento da lista de links externos.
sleep(5)
```

```python
# ==============================================================================
# 6. LOCALIZAÇÃO AVANÇADA (XPATH) E CAPTURA DA REDE SOCIAL
# ==============================================================================

# 'find_element' (no singular) localiza apenas o primeiro elemento que atende ao critério.
# Usamos uma expressão XPATH (XML Path Language) que varre a árvore do HTML procurando:
# '//a' -> Qualquer tag de link (âncora) na página...
# '[contains(@href, "instagram.com")]' -> Onde o atributo de endereço web (href) contenha o texto "instagram.com".
# '.get_attribute("href")' extrai a URL completa contida dentro desse link localizado.
act_url = driver.find_element(By.XPATH, '//a[contains(@href, "instagram.com")]').get_attribute("href")

# Demonstração de tratamento da URL obtida (Exemplo: "https://www.instagram.com/wagnermouraofficial/"):
# 1. '.strip("/")' remove barras residuais do início ou final do texto.
# 2. '.strip()' remove espaços vazios acidentais.
# 3. '.split("/")' quebra a URL pelas barras para que o analista possa isolar o arroba/username no índice correto.
act_url.strip("/").strip().split("/")

# Aguarda 5 segundos antes de encerrar o bloco de execução.
sleep(5)
```
